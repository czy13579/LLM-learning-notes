```python
from http import HTTPStatus
import dashscope
def sample_call_streaming():
    prompt_text = 'ç”¨èåœã€åœŸè±†ã€èŒ„å­åšé¥­ï¼Œç»™æˆ‘ä¸ªèœè°±ã€‚'
    response_generator = dashscope.Generation.call(
        model='qwen-turbo',
        prompt=prompt_text,
        # è®¾ç½®streamä¸ºTrueï¼Œå¼€å¯æµå¼è¾“å‡º
        #api_key=my_qwen_key,
        stream=True,
        incremental_output=True,
        top_p=0.8)

    for response in response_generator:
        if response.status_code == HTTPStatus.OK:
            print(response.output.text, end="")  # è¾“å‡ºæ–‡æœ¬
        else:
            print(response.code)  # é”™è¯¯ç 
            print(response.message)  # é”™è¯¯ä¿¡æ¯


if __name__ == '__main__':
    sample_call_streaming()
```
